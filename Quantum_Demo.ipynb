```python
print("Hello, Quantum World!")

import cirq

# Create two qubits
q0, q1 = cirq.LineQubit.range(2)

# Build the Bell state circuit
circuit = cirq.Circuit(
    cirq.H(q0),
    cirq.CNOT(q0, q1),
    cirq.measure(q0, q1, key="result")   # <-- REQUIRED
)

# Run the simulation
sim = cirq.Simulator()
result = sim.run(circuit, repetitions=1000)

print("Circuit:")
print(circuit)
print("\nResults:")
print(result)

```

Grover's Algorithm

```python
import cirq

# Define 2 qubits
q = cirq.LineQubit.range(2)

# Oracle that marks |11>
def oracle():
    return [cirq.Z(q[0]).controlled_by(q[1])]

# Diffuser (inversion about the mean)
def diffuser():
    return [
        cirq.H.on_each(*q),
        cirq.X.on_each(*q),
        cirq.H(q[1]).controlled_by(q[0]),
        cirq.X.on_each(*q),
        cirq.H.on_each(*q)
    ]

# Build the Grover circuit
circuit = cirq.Circuit()

# Step 1: Superposition
circuit.append(cirq.H.on_each(*q))

# Step 2: Oracle
circuit.append(oracle())

# Step 3: Diffuser
circuit.append(diffuser())

# ğŸ‘‰ REQUIRED IN NEW CIRQ (measurement)
circuit.append(cirq.measure(*q, key="result"))

# Run the circuit
sim = cirq.Simulator()
result = sim.run(circuit, repetitions=1000)

print("Grover Circuit:")
print(circuit)
print("\nResults:")
print(result)
```
